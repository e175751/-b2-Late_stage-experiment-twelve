```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = 1.50
hoge = []
for i in range(0,101):
    init = r * (1 - init) * init
    hoge.append(init)
print(hoge)
```

```python
def logistic(r):
    init = 0.7
    element = []
    #element.append(init)
    for i in range(1,100):
        element.append(init)
        init = round(r * (1 - init) * init , 3)
    return element
```

```python
print(logistic(1.50))
```

```python
#x = np.arange(0,100)
data = np.array(logistic(1.50))
plt.plot(data)
plt.show()
```

```python
path_w = "test1.txt"
#l_n_str = [str(n) for n in l_n]
data = [str(i) for i in logistic(1.50)]
with open(path_w, mode='w') as f:
    f.writelines('\n'.join(data))
     #f.write("\n")
    #f.write(data)

with open(path_w) as f:
    print(f.read())
```
